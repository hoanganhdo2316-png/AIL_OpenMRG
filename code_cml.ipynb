```python
import pandas as pd
import netCDF4 as nc
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

```python
ds = xr.open_dataset('cml.nc')
df = ds.isel(time=slice(0, 100)).to_dataframe()
df
```

```python
df_2=pd.read_csv('cml_metadata.csv')
df_2[df_2['Sublink']==1]
```

```python
df_combined=pd.merge(
    df,
    df_2,
    left_on='sublink',
    right_on='Sublink',
    how='inner'
)
```

```python
df=df.reset_index()
df.index.names
```
